```python
# creating new dataframe to manipulate
cleaned_df = df

# replacing missing values in Cost
# column with mean of that column
cleaned_df['Cost'] = cleaned_df['Cost'].fillna(cleaned_df['Cost'].mean())

# replacing missing values in Mass (g) column
# with median of that column
cleaned_df['Mass (g)'] = cleaned_df['Mass (g)'].fillna(cleaned_df['Mass (g)'].median())

# replacing missing values in Length column
# with mode of that column
cleaned_df['Length'] = cleaned_df['Length'].fillna(cleaned_df['Length'].mode()[0])

# replacing missing values in Density (g/mL) column
# with my own value (0.60)
cleaned_df['Density (g/mL)'] = cleaned_df['Density (g/mL)'].fillna(0.60)

# replacing missing values in Google column
# with backward filling of that column
cleaned_df['Google'] = cleaned_df['Google'].fillna(method='bfill')

# replacing missing values in Yelp column
# with forwards filling of that column
cleaned_df['Yelp'] = cleaned_df['Yelp'].fillna(method='ffill')


cleaned_df
```
